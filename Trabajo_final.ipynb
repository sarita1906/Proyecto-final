```python
import pandas as pd
import numpy as np
import altair as alt
import wbgapi as wb
```

```python
#Llamamos al Dataframe que se usará
development = pd.read_csv("Global Development Data.csv")
development
```

```python
#Para saber las categorías de Series que están incluidas en esta base
development.groupby('Series Name')['Series Code'].unique()
```

```python
#Para saber sobre aquellas economías incluidas en la base
lst = wb.economy.info().table()
lst
```

```python
#Se crea un Dataframe para ordenar un poco utilizando lista de listas
nombres = pd.DataFrame(data = lst,
                       columns = ["economy", "Country", "Region", "income_classification"])
nombres.head(3)
```

```python
#Tomando en cuenta las categorias de series name :1. Electricidad
tb1 = wb.data.DataFrame('EG.ELC.ACCS.ZS',
                  time = range(2010, 2022), labels = True)

tb1 = tb1.reset_index()

tb1.rename(columns = {'EG.ELC.ACCS.ZS': 'electricidad'}, inplace = True)



tb1_f = pd.merge(tb1,
                     nombres,
                     on = ["economy", "Country"],
                     how = "inner")
```

```python
#Tomando en cuenta las categorias de series name :2. Electricidad rural
tb2 = wb.data.DataFrame('EG.ELC.ACCS.RU.ZS',
                  time = range(2010, 2022), labels = True)

tb2 = tb2.reset_index()

tb2.rename(columns = {'EG.ELC.ACCS.RU.ZS': 'electricidad_rural'}, inplace = True)



tb2_f = pd.merge(tb2,
                     nombres,
                     on = ["economy", "Country"],
                     how = "inner")
```

```python
#Tomando en cuenta las categorias de series name :3. Emisión de CO2
tb3 = wb.data.DataFrame('EN.ATM.CO2E.PC',
                  time = range(2010, 2022), labels = True)

tb3 = tb3.reset_index()

tb3.rename(columns = {'EN.ATM.CO2E.PC': 'emision_co2'}, inplace = True)



tb3_f = pd.merge(tb3,
                     nombres,
                     on = ["economy", "Country"],
                     how = "inner")
```

```python
#Tomando en cuenta las categorias de series name :4. Crecimiento de PBI
tb4 = wb.data.DataFrame('NY.GDP.MKTP.KD.ZG',
                  time = range(2010, 2022), labels = True)

tb4 = tb4.reset_index()

tb4.rename(columns = {'NY.GDP.MKTP.KD.ZG': 'pbi'}, inplace = True)



tb4_f = pd.merge(tb4,
                     nombres,
                     on = ["economy", "Country"],
                     how = "inner")
```

```python
#Tomando en cuenta las categorias de series name :5. Camas de hospital
tb5 = wb.data.DataFrame('SH.MED.BEDS.ZS',
                  time = range(2010, 2022), labels = True)

tb5 = tb5.reset_index()

tb5.rename(columns = {'SH.MED.BEDS.ZS': 'camas_hospital'}, inplace = True)



tb5_f = pd.merge(tb5,
                     nombres,
                     on = ["economy", "Country"],
                     how = "inner")
```

```python
#Tomando en cuenta las categorias de series name :6. Fuerza laboral femenina
tb6 = wb.data.DataFrame('SH.MED.BEDS.ZS',
                  time = range(2010, 2022), labels = True)

tb6 = tb6.reset_index()

tb6.rename(columns = {'SL.TLF.TOTL.FE.ZS': 'labor_femenina'}, inplace = True)



tb6_f = pd.merge(tb6,
                     nombres,
                     on = ["economy", "Country"],
                     how = "inner")
```

```python
# Eliminando aquellos que tengan como Region Missing values
dataframes = [tb1_f, tb2_f, tb3_f, tb4_f, tb5_f, tb6_f]

for df in dataframes:
    df = df.loc[df['Region'] != '']
```

```python
#Conviertiendo a formato long (también se puede realizar con loops, lo probe y funciona pero aqui esta desagregado)

tb1_melt = tb1_f.melt(id_vars = ['economy', 'Country', 'Region', 'income_classification'],
                   value_vars = [col for col in tb1 if col.startswith('YR')])
tb2_melt = tb2_f.melt(id_vars = ['economy', 'Country', 'Region', 'income_classification'],
                   value_vars = [col for col in tb2 if col.startswith('YR')])
tb3_melt = tb3_f.melt(id_vars = ['economy', 'Country', 'Region', 'income_classification'],
                   value_vars = [col for col in tb3 if col.startswith('YR')])
tb4_melt = tb4_f.melt(id_vars = ['economy', 'Country', 'Region', 'income_classification'],
                   value_vars = [col for col in tb4 if col.startswith('YR')])
tb5_melt = tb5_f.melt(id_vars = ['economy', 'Country', 'Region', 'income_classification'],
                   value_vars = [col for col in tb5 if col.startswith('YR')])
tb6_melt = tb6_f.melt(id_vars = ['economy', 'Country', 'Region', 'income_classification'],
                   value_vars = [col for col in tb6 if col.startswith('YR')])
```

```python
#Cambiar YR de las observaciones para que solo sea un numérico (también se puede realizar con loops, lo probe y funciona pero aqui esta desagregado)
tb1_melt['year'] = tb1_melt['variable'].str.replace('YR', '').astype('int')
tb2_melt['year'] = tb2_melt['variable'].str.replace('YR', '').astype('int')
tb3_melt['year'] = tb3_melt['variable'].str.replace('YR', '').astype('int')
tb4_melt['year'] = tb4_melt['variable'].str.replace('YR', '').astype('int')
tb5_melt['year'] = tb5_melt['variable'].str.replace('YR', '').astype('int')
tb6_melt['year'] = tb6_melt['variable'].str.replace('YR', '').astype('int')
```
